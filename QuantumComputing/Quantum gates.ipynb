```python
qc = QuantumCircuit(1,1)
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
qc.x(0)
qc.measure(0,0)
counts = execute(qc, Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

```python
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(3,3)
```

```python
qc.x(0)
qc.measure(0,0)
qc.h(1)
qc.measure(1,1)
qc.z(0)
qc.measure(0,1)
qc.cx(0,1)
qc.measure([0,1],[0,1])
qc.x(1)
qc.cz(1,2)
qc.measure([1,2],[1,2])
```

```python
qc.draw(output='mpl')
```
