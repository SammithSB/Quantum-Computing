```python
from qiskit import *
IBMQ.load_account()
provider = IBMQ.get_provider()
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
qc = QuantumCircuit(3,1)
```

```python

qc.h(1)
qc.cx(1,2)
qc.barrier()
qc.draw(output='mpl')
```

```python

qc.draw(output='mpl')
```

```python

qc.cx(0,1)
qc.h(0)
qc.draw(output='mpl')
```

```python
qc.barrier()

qc.barrier()
qc.cx(1,2)
qc.cz(0,2)
qc.barrier()
qc.measure(2,0)
qc.draw(output='mpl')
```

```python
counts = execute(qc, Aer.get_backend('qasm_simulator')).result().get_counts()
```

```python
plot_histogram(counts)
```

```python
count = execute(qc, Aer.get_backend('statevector_simulator')).result().get_statevector()
```

```python
from qiskit.tools.visualization import plot_bloch_multivector
```

```python
plot_bloch_multivector(count)
```

```python
from qiskit.tools import job_monitor
```

```python
job = execute(qc, provider.get_backend('ibmq_burlington'))
```

```python
job_monitor(job)
```

```python
job = job.result().get_counts()
```

```python
plot_histogram(job)
```
